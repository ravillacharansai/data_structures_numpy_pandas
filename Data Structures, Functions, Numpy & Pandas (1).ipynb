## Section C: Exploring Pandas

```python
wa = {
    'name': ['sai', 'charan', 'balu', 'thishan', 'giri',
             'sasi', 'malli', 'viyaan', 'sreenu', 'ramu'],
    
    'age': [21, 22, 33, 44, 20, 55, 23, 32, 25, 30],
    
    'city': ['chennai', 'pune', 'tirupati', 'kadapa', 'hyderabad',
             'mumbai', 'annamaya', 'vizag', 'kakinada', 'bengaluru'],
    
    'salary': [20000, 30000, 40000, 24000, 33000,
               50000, 60000, 10000, 45000, 76000]
}
```

```python
sa=pd.DataFrame(wa)
sa
```

```python
sa.info()
```

```python
sa.describe()
```

```python
sa[['name','city']]
```

```python
sa.drop(columns=['city'])
```

```python
salary_mean=sa['salary'].mean()
sa['salary']=sa['salary'].fillna(salary_mean)
print(sa)
```
